```python
cleaned_6_9_df['Major'] = cleaned_6_9_df['Major'].str.replace('.\n', ' ').str.rstrip('. ')
cleaned_6_9_df['Major']
```

#### Create a list of valid majors

```python
valid_majors = colleges123['MAJORS'].tolist()
valid_majors
```

```python
# Define a function to find the closest matching valid major
def get_closest_valid_major(major, valid_majors):
    match, score = process.extractOne(major, valid_majors)
    if score >= 95:  # Adjust the threshold as needed
        return match
    else:
        return major
    # return process.extractOne(major, valid_majors)[0]

# Apply the function to each row in df1_2['Major']
cleaned_6_9_df['CleanedMajor'] = cleaned_6_9_df['Major'].apply(lambda x: get_closest_valid_major(x, valid_majors))
```

```python
cleaned_6_9_df[['Major', 'CleanedMajor']]
```

```python
cleaned_6_9_df['InvalidMajor'] = None  # Initialize with None
```

```python
for index, row in cleaned_6_9_df.iterrows():
    major = row['CleanedMajor']
    # Adjust the threshold as needed
    if process.extractOne(major, valid_majors)[1] < 95:
        # Mark as invalid by storing in the 'InvalidMajor' column
        cleaned_6_9_df.at[index, 'InvalidMajor'] = major
```

```python
cleaned_6_9_df[['Major', 'CleanedMajor','InvalidMajor']]
```

```python
invalid_major_mapping = {
    'B Admin': 'Business Admin',
    'Info Sc': 'Information Sc',
    'CS': 'Computer Sc',
    'BusinessAD': 'Business Admin',
    'Biol': 'Biology',
    'Chem': 'Chemistry',
    'Element': 'Elementary Ed',
    'Elementary.': 'Elementary Ed',
    'Secondary.': 'Secondary Ed',
    'MBA': 'Business Administration',
    'Engl': 'English',
}
```

Create a mapping from the invalid majors

Backpropogate to the 'CleanedMajor' Column to replace all found invalid majors, with the correct major defined in the invalid_major_mapping dictionary. 

```python
for index, row in cleaned_6_9_df.iterrows():
    major = row['CleanedMajor']
    if major in invalid_major_mapping:
        cleaned_6_9_df.at[index, 'CleanedMajor'] = invalid_major_mapping[major]
```

```python
unique_cities = cleaned_7_9_df['Address'].str.extract(r'([A-Za-z\s]+)', expand=False).unique()
unique_cities
```

```python
valid_cities = ['Denver', 'Pine Bluff', 'Little Rock', 'Atlanta', 'Conway',
       'Savannah', 'Seattle', 'Dallas', 'Orlando', 'New York']
```

```python
# Define a function to find the closest matching valid major
def get_closest_valid_cities(city, valid_cities):
    match, score = process.extractOne(city, valid_cities)
    if score >= 60:  # Adjust the threshold as needed
        return match
    else:
        return city
    # return process.extractOne(major, valid_majors)[0]

# Apply the function to each row in df1_2['Major']
cleaned_7_9_df['CleanedAddress'] = cleaned_7_9_df['Address'].apply(lambda x: get_closest_valid_cities(x, valid_cities))
```

```python
cleaned_7_9_df['InvalidCity'] = None  # Initialize with None

for index, row in cleaned_7_9_df.iterrows():
    city = row['CleanedAddress']
    # Adjust the threshold as needed
    if process.extractOne(city, valid_cities)[1] < 95:
        # Mark as invalid by storing in the 'InvalidCity' column
        cleaned_7_9_df.at[index, 'InvalidCity'] = city
        
cleaned_7_9_df[['Address', 'CleanedAddress','InvalidCity']]
```
