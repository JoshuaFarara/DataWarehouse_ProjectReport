```python
# Map categorical values to their corresponding IDs
pop_Integrated_Data = pop_Integrated_Data.merge(dim_degrees, on=['MajorID', 'DegreeID'], how='left')
pop_Integrated_Data = pop_Integrated_Data.merge(dim_time, on=['YearID', 'SemesterID'], how='left')
pop_Integrated_Data = pop_Integrated_Data.merge(dim_status, on='StatusID', how='left')
pop_Integrated_Data = pop_Integrated_Data.merge(dim_gparank, on='GPARankID', how='left')
```
